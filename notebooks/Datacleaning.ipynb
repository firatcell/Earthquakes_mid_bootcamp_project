```python
data = data.drop(columns=['Search Parameters'])
```

```python
data.isna().count()
```

```python
data['Year'] = data['Year'].fillna(0)
data['Mo'] = data['Mo'].fillna(0)
data['Dy'] = data['Dy'].fillna(0)
data['Hr'] = data['Hr'].fillna(0)
data['Mn'] = data['Mn'].fillna(0)
data['Sec'] = data['Sec'].fillna(0)
```

```python
data['Datetime'] = pd.to_datetime(data['Year'].astype(int).astype(str) + '-' +
                                  data['Mo'].astype(int).astype(str) + '-' +
                                  data['Dy'].astype(int).astype(str) + ' ' +
                                  data['Hr'].astype(int).astype(str) + ':' +
                                  data['Mn'].astype(int).astype(str) + ':' +
                                  data['Sec'].astype(int).astype(str),
                                  errors='coerce')
```

```python
data = data.drop(columns=['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec'])
```

```python
tsu_na_counts = data['Tsu'].isna().value_counts()
tsu_na_counts
```

```python
def convert_to_boolean(val):
    if pd.isna(val):
        return False
    else:
        return True
```

```python
data['Tsu'] = data['Tsu'].apply(convert_to_boolean)
```

```python
data['Tsu'].value_counts()
```

```python
vol_na_counts = data['Vol'].isna().value_counts()
vol_na_counts
```

```python
data['Vol'] = data['Vol'].apply(convert_to_boolean)
```

```python
# select columns containing 'Description' in their column name
selected_cols = data.filter(regex='^(?!.*\(\$Mil\)).*(Description|Total)')

# convert the selected columns to int
converted_cols = selected_cols.astype('int')

# replace the original columns with the converted columns in the original dataframe
data[converted_cols.columns] = converted_cols

# convert 'MMI Int' and 'Deaths' columns to int
data['MMI Int'] = data['MMI Int'].astype(int)
data['Deaths'] = data['Deaths'].astype(int)
```
